```python
def extract_numerical_values(text):
    # Regular expression pattern to extract numerical values (including billions)
    pattern = r'\b(?:\d+\.)?\d+(?: billion)?\b'

    # Find all matches of the pattern in the text
    matches = re.findall(pattern, text)

    # Convert the matched strings to floats
    numerical_values = [float(match.replace(' billion', '')) * 1e9 if 'billion' in match else float(match) for match in matches]

    return numerical_values
```

```python
import numpy as np

# Historical revenue data (for simplicity, using the provided revenue values)
historical_revenue = np.array([data['Revenue'], data['Total Revenue Expected']])

# Exponential smoothing parameters
alpha = 0.2  # Smoothing parameter (0 < alpha < 1)

# Perform exponential smoothing
smoothed_values = [historical_revenue[0]]
for i in range(1, len(historical_revenue)):
    smoothed_value = alpha * historical_revenue[i] + (1 - alpha) * smoothed_values[-1]
    smoothed_values.append(smoothed_value)

# Forecast future revenue using exponential smoothing
forecasted_values = [smoothed_values[-1]]
for i in range(1, 3):  # Forecasting for the next 2 years
    forecasted_value = alpha * forecasted_values[-1] + (1 - alpha) * smoothed_values[-1]
    forecasted_values.append(forecasted_value)

# Plotting historical revenue and forecasted revenue
years = ['2023', '2024', '2025', '2026']
plt.figure(figsize=(10, 6))
plt.plot(years[:2], historical_revenue, marker='o', label='Historical Revenue', color='blue')
plt.plot(years[1:], forecasted_values, marker='o', label='Forecasted Revenue', color='red')
plt.title('Historical and Forecasted Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue (in billions)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define scenario parameters
revenue_increase_percent = 10  # 10% increase in revenue
expenditure_decrease_percent = 5  # 5% decrease in expenditure

# Calculate scenario-based financial metrics
scenario_revenue = data['Total Revenue Expected'] * (1 + revenue_increase_percent / 100)
scenario_expenditure = data['Total Expenditure'] * (1 - expenditure_decrease_percent / 100)

# Calculate key financial ratios for the scenario
scenario_current_ratio = scenario_revenue / scenario_expenditure
scenario_profit_margin = (scenario_revenue - scenario_expenditure) / scenario_revenue
scenario_debt_to_equity_ratio = data['Total Government Debt'] / scenario_expenditure

# Display the scenario-based financial metrics
print("Scenario Analysis Results:")
print("Scenario Revenue:", scenario_revenue, "billion")
print("Scenario Expenditure:", scenario_expenditure, "billion")
print("Scenario Current Ratio:", scenario_current_ratio)
print("Scenario Profit Margin:", scenario_profit_margin)
print("Scenario Debt-to-Equity Ratio:", scenario_debt_to_equity_ratio)
```

```python
# Define range of revenue growth rates for sensitivity analysis
revenue_growth_rates = np.linspace(-0.2, 0.2, 11)  # From -20% to +20% in steps of 4%

# Calculate profit margin and return on equity for each revenue growth rate
profit_margins = []
return_on_equities = []

for growth_rate in revenue_growth_rates:
    scenario_revenue = data['Total Revenue Expected'] * (1 + growth_rate)
    scenario_expenditure = data['Total Expenditure']
    scenario_profit = scenario_revenue - scenario_expenditure
    profit_margin = scenario_profit / scenario_revenue
    return_on_equity = scenario_profit / (data['Total Government Debt'] + scenario_expenditure)

    profit_margins.append(profit_margin)
    return_on_equities.append(return_on_equity)

# Plot sensitivity analysis results
plt.figure(figsize=(10, 6))
plt.plot(revenue_growth_rates, profit_margins, marker='o', label='Profit Margin', color='blue')
plt.plot(revenue_growth_rates, return_on_equities, marker='o', label='Return on Equity', color='red')
plt.title('Sensitivity Analysis: Revenue Growth Rate')
plt.xlabel('Revenue Growth Rate')
plt.ylabel('Ratio')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Expenditure Efficiency Analysis

# 1. Calculate the efficiency of government spending by comparing budget allocation to actual performance and outcomes.

# Define functions to calculate efficiency metrics
def calculate_efficiency(budget_allocation, actual_performance):
    """
    Calculate expenditure efficiency.

    Parameters:
    - budget_allocation (float): Budget allocation for a specific program or department.
    - actual_performance (float): Actual performance or outcomes achieved.

    Returns:
    - efficiency (float): Expenditure efficiency ratio.
    """
    efficiency = actual_performance / budget_allocation
    return efficiency

# Example usage
budget_allocation = 1000000  # Budget allocation for a program
actual_performance = 850000  # Actual outcomes achieved
efficiency_ratio = calculate_efficiency(budget_allocation, actual_performance)
print("Expenditure Efficiency Ratio:", efficiency_ratio)

# 2. Evaluate the effectiveness of expenditures in achieving policy objectives, delivering public services, and addressing key priorities.

# Define functions to evaluate effectiveness
def evaluate_effectiveness(policy_objectives, public_services, key_priorities):
    """
    Evaluate the effectiveness of expenditures.

    Parameters:
    - policy_objectives (list): List of policy objectives.
    - public_services (list): List of public services delivered.
    - key_priorities (list): List of key priorities addressed.

    Returns:
    - effectiveness_score (int): Effectiveness score based on the evaluation.
    """
    # Example: Calculate effectiveness score based on the number of objectives, services, and priorities addressed
    effectiveness_score = len(policy_objectives) + len(public_services) + len(key_priorities)
    return effectiveness_score

# Example usage
policy_objectives = ["Reduce unemployment", "Improve healthcare access"]
public_services = ["Healthcare services", "Education programs"]
key_priorities = ["Infrastructure development", "Poverty alleviation"]
effectiveness_score = evaluate_effectiveness(policy_objectives, public_services, key_priorities)
print("Effectiveness Score:", effectiveness_score)

# 3. Identify areas where expenditures may be streamlined, reallocated, or optimized to enhance efficiency and maximize value for money.

# Provide recommendations based on analysis

def expenditure_recommendations(inefficient_programs, reallocation_opportunities, optimization_strategies):
    """
    Provide recommendations for expenditure optimization.

    Parameters:
    - inefficient_programs (list): List of inefficient programs or departments.
    - reallocation_opportunities (dict): Dictionary containing opportunities for budget reallocation.
    - optimization_strategies (list): List of recommended optimization strategies.

    Returns:
    - recommendations (dict): Dictionary containing recommendations.
    """
    recommendations = {
        "inefficient_programs": inefficient_programs,
        "reallocation_opportunities": reallocation_opportunities,
        "optimization_strategies": optimization_strategies
    }
    return recommendations

# Example usage
inefficient_programs = ["Program A", "Department B"]
reallocation_opportunities = {"Program C": "Allocate additional funds"}
optimization_strategies = ["Implement cost-saving measures", "Streamline administrative processes"]
recommendations = expenditure_recommendations(inefficient_programs, reallocation_opportunities, optimization_strategies)
print("Expenditure Optimization Recommendations:", recommendations)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

# Provided text
text = """
71. Mr Speaker, the Trade sector is one of the identified priority sectors to promote
economic growth. The sector has been allocated M599.5 million. A significant
progress has been achieved with the construction completion of 16 and 7 factory
shells and associated infrastructure at Ha Belo and Tikoe marking a major
milestone. This infrastructure supports and elevates our relentless efforts to
market Lesotho as a favourable investment destination in Southern Africa.
72. The Lesotho National Development Corporation has this year received firm
indications for new investments to utilize this infrastructure amounting to 696
million. The areas of interest have been across sectors but mostly in textiles,
clothing, stainless steel bars, electrical wiring accessories, leather seat covers,
laundry services, fabric dying and garment manufacturing. This makes
investment prospects over the medium term very bright. We are positive that
these investments will be actualized, and considerable jobs will be created. To
be precise, these investment opportunities are poised to register over 5,500 jobs.
73. Mr. Speaker, limited capacity and access to affordable finance continue to be
constraints affecting MSME growth. To address this challenge, government is
enhancing capacity and providing funding to targeted groups through
empowerment programmes such as Lesotho Youth Acceleration Project,
Competitiveness and Financial Inclusion Project and Lesotho Enterprises
Assistance Programme.
"""

# Function to extract sector and budget information
def extract_sector_budget(text):
    sectors = []
    budgets = []
    lines = text.split('\n')
    for line in lines:
        if line.strip().startswith(('71.', '72.', '73.', '74.', '75.', '76.', '77.', '78.', '79.', '80.', '81.', '82.', '83.', '84.', '85.', '86.', '87.', '88.', '89.', '90.', '91.', '92.', '93.', '94.', '95.', '96.', '97.', '98.', '99.', '100.', '101.', '102.', '103.', '104.', '105.', '106.', '107.', '108.', '109.', '110.', '111.', '112.', '113.', '114.', '115.', '116.', '117.', '118.', '119.', '120.', '121.', '122.', '123.')):
            sector = line.split(',')[0].strip()
            budget_str = line.split('M')[-1].split(' ')[0]
            budget_str = ''.join(filter(str.isdigit, budget_str))  # Remove non-numeric characters
            if budget_str:  # Check if string is not empty
                budgets.append(float(budget_str))
                sectors.append(sector)
    return sectors, budgets

# Extracting sector and budget information
sectors, budgets = extract_sector_budget(text)

# Plotting budget allocation
plt.figure(figsize=(10, 8))
plt.barh(sectors, budgets, color='skyblue')
plt.xlabel('Budget Allocation (Million M)')
plt.ylabel('Sectors')
plt.title('Budget Allocation for Different Sectors')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()

# Generating word cloud for the text
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plotting word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Provided Text')
plt.show()
```
